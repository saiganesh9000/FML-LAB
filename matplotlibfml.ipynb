# **Scatter Plot**

It is used to represent a pair of numerical data one with x-axis and another with y-axis.each representation will be in dot form

```python
x = [2,4,6,8,10,12]
y = [50,70,20,80,90,40]
plt.scatter(x,y) #it relates x with y (representation in the form of plot)
plt.show()
```

```python
x = [2,4,6,8,10,12]
y = [50,70,20,80,90,40]
c = ["orange","yellow","purple","black","green","maroon"]
plt.scatter(x,y , c= c) #it relates x with y (representation in the form of plot)
plt.show()
```

```python
x = [2,4,6,8,10,12]
y = [50,70,20,80,90,40]
#c = ["orange","yellow","purple","black","green","maroon"]
plt.scatter(x,y , color = "maroon" ) #it relates x with y (representation in the form of plot)
plt.show()
```

```python
x = [2,4,6,8,10,12]
y = [50,70,20,80,90,40]
#c = ["orange","yellow","purple","black","green","maroon"]
size = [100,200,50,150,300,250]
plt.scatter(x,y , color = "maroon" , s = size ) #it relates x with y (representation in the form of plot)
plt.show()
```

# **Histogram**

Histogram is accurate representation of the distribution of numerical data

Divide the data into different intervals

```python
marks = [90,95,20,35,70,75,60,65,30,55]
grade_intervals = [0,50,80,100]
plt.title("Student Grade")
plt.hist(marks,grade_intervals,histtype = "bar" , rwidth = 0.7) # reduce width
plt.xticks([0,50,80,100])
plt.xlabel("percentage")
plt.ylabel("No.of Students")
plt.show()
```

```python
marks = [90,95,20,35,70,75,60,65,30,55]
grade_intervals = [0,50,80,100]
plt.title("Student Grade")
plt.hist(marks,grade_intervals,histtype = "step" , rwidth = 0.7) # reduce width
plt.xticks([0,50,80,100])
plt.xlabel("percentage")
plt.ylabel("No.of Students")
plt.show()
```

```python
marks = [90,95,20,35,70,75,60,65,30,55]
grade_intervals = [0,50,80,100]
plt.title("Student Grade")
plt.hist(marks,grade_intervals,histtype = "stepfilled" , rwidth = 0.7 , facecolor = "r") # reduce width
plt.xticks([0,50,80,100])
plt.xlabel("percentage")
plt.ylabel("No.of Students")
plt.show()
```

```python
marks = [90,95,20,35,70,75,60,65,30,55]
grade_intervals = [0,50,80,100]
plt.title("Student Grade")
plt.hist(marks,grade_intervals,histtype = "bar" , rwidth = 0.7 , facecolor = "g") # reduce width
plt.xticks([0,50,80,100])
plt.xlabel("percentage")
plt.ylabel("No.of Students")
plt.show()
```

# **BOX PLOT**

> The shape of the boxplot shows how the dta is distributed and it also shows any outliers

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
list1 = [11,4,6,8,6,9,3]
fig = plt.figure(figsize = (10,7))
plt.boxplot(list1)
plt.show()
```

```python
list2 = [1,2,2,2,3,5,6]
fig = plt.figure(figsize = (10,7))
plt.boxplot(list2)
plt.show()
```

```python
list1 = [11,4,6,8,6,9,3]
fig = plt.figure(figsize = (5,5))
plt.boxplot(list1 , notch= True)
plt.show()
```

```python
list1 = [11,4,6,8,6,9,3]
fig = plt.figure(figsize = (10,7))
plt.boxplot(list1 , vert= True)
plt.show()
```

```python
list1 = [11,4,6,8,6,9,3]
fig = plt.figure(figsize = (5,4))
plt.boxplot(list1 , widths = 0.6)
plt.show()
```

```python
list1 = [11,4,6,8,6,9,3]
fig = plt.figure(figsize = (5,4))
plt.boxplot(list1 , labels = ["ssl"] , widths = 0.5)
plt.show()
```

```python
list1 = [11,4,6,8,6,9,3]
fig = plt.figure(figsize = (5,4))
plt.boxplot(list1 , labels = ["SSL"] , widths = 0.3 , patch_artist = True)
plt.show()
```

```python
list1 = [11,4,6,8,6,9,3]
fig = plt.figure(figsize = (5,4))
plt.boxplot(list1 , labels = ["SSL"] , widths = 0.3 , patch_artist = False, showmeans = True)
plt.show()
```

```python
list1 = [11,4,6,8,6,9,3]
fig = plt.figure(figsize = (5,4))
plt.boxplot(list1 , labels = ["SSL"] , widths = 0.3 , patch_artist = False, showmeans = True , boxprops = dict(color="LIME"))
plt.show()
```

# **PIECHART**

> A Pie chart is a circular statistical plot that can display only one series of data. The area of chart is the total percentage of given data



```python
import matplotlib.pyplot as plt
```

```python
Student_performance = ['Excellent' , 'Good' , 'Average' , 'Poor']
Student_values = [35,30,20,15]
plt.pie(Student_values)
plt.show()
```

```python
Student_performance = ['Excellent' , 'Good' , 'Average' , 'Poor']
Student_values = [35,30,20,15]
plt.pie(Student_values , labels = Student_performance)
plt.show()
```

```python
Student_performance = ['Excellent' , 'Good' , 'Average' , 'Poor']
Student_values = [35,30,20,50]
plt.pie(Student_values , labels = Student_performance , startangle = 180)
# startangle means where it should represents
plt.show()
```

```python
Student_performance = ['Excellent' , 'Good' , 'Average' , 'Poor']
Student_values = [35,30,20,15]
plt.pie(Student_values , labels = Student_performance , startangle = 90)
# startangle means where it should represents
# counter plots
plt.show()
```

```python
Student_performance = ['Excellent' , 'Good' , 'Average' , 'Poor']
Student_values = [35,30,20,15]
plt.pie(Student_values , labels = Student_performance , startangle = 0 , explode = [0.3,0.3,0.3,0.3])
# startangle means where it should represents
# counter plots
# explode extract takes default value 0 , if you give any value it seperates from the pie chart
plt.show()
```

```python
Student_performance = ['Excellent' , 'Good' , 'Average' , 'Poor']
Student_values = [35,30,20,15]
plt.pie(Student_values , labels = Student_performance , startangle = 0 , explode = [0.2,0.2,0.2,0.2])
# startangle means where it should represents
# counter plots
# explode extract takes default value 0 , if you give any value it seperates from the pie chart
plt.show()
```

```python
Student_performance = ['Excellent' , 'Good' , 'Average' , 'Poor']
Student_values = [35,30,20,15]
plt.pie(Student_values , labels = Student_performance , startangle = 0 , explode = [0.2,0.2,0.2,0.2],shadow = True)
# startangle means where it should represents
# counter plots
# explode extract takes default value 0 , if you give any value it seperates from the pie chart
plt.show()
```

```python
Student_performance = ['Excellent' , 'Good' , 'Average' , 'Poor']
Student_values = [35,30,20,15]
plt.pie(Student_values , labels = Student_performance , startangle = 0 , explode = [0.2,0.2,0.2,0.2] , colors = ["pink" , "yellow" , "purple" , "skyblue"])
# startangle means where it should represents
# counter plots
# explode extract takes default value 0 , if you give any value it seperates from the pie chart
plt.show()
```

```python
Student_performance = ['Excellent' , 'Good' , 'Average' , 'Poor']
Student_values = [35,30,20,15]
plt.pie(Student_values , labels = Student_performance , startangle = 0 , explode = [0.2,0.2,0.2,0.2] , colors = ["pink" , "yellow" , "purple" , "skyblue"])
# startangle means where it should represents
# counter plots
# explode extract takes default value 0 , if you give any value it seperates from the pie chart
plt.legend(title = "Student_performance")
plt.show()
```

```python
Student_performance = ['Excellent' , 'Good' , 'Average' , 'Poor']
Student_values = [35,30,20,15]
plt.pie(Student_values , labels = Student_performance , startangle = 0 , explode = [0.2,0.2,0.2,0.2] , colors = ["pink" , "yellow" , "purple" , "skyblue"] , autopct = "%2.2f%%")
# startangle means where it should represents
# counter plots
# explode extract takes default value 0 , if you give any value it seperates from the pie chart
plt.show()
```
